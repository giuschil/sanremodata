### 1 tabella dati sui festival da 1951 a 2023

```python
# Lista per memorizzare i dati estratti da tutte le pagine
data_list = []
year_start = 1951
year_end = 2024

# Ciclo attraverso gli anni da 1951 a 2023
for year in range(year_start, year_end):
    # Costruisci l'URL per l'anno corrente
    url = f'https://it.wikipedia.org/wiki/Festival_di_Sanremo_{year}'

    # Effettua la richiesta HTTP per ottenere il contenuto della pagina
    response = requests.get(url)
    if response.status_code == 200:
        # Utilizza BeautifulSoup per analizzare l'HTML della pagina
        soup = BeautifulSoup(response.content, 'html.parser')

        # Trova tutti gli elementi <th>
        period_elements = soup.find('th', text='Periodo')
        sede_elements = soup.find('th', text='Sede')
        speaker_elements = soup.find('th', text='Presentatore')
        singers_elements = soup.find('th', text='Partecipanti')
        winner_elements = soup.find('th', text='Vincitore')
        # Estrai il titolo della sede associato a ciascun elemento <th> trovato
        for sede_element in sede_elements:
            # Ottieni il titolo della sede
            if period_elements is not None:
                #singers = singers_elements.get_text(strip=True)
                period = period_elements.find_next('td').get_text(strip=True)
                # Esegui altre operazioni con il testo estratto
            else:
                print(year," - ","Elemento non trovato.")
            sede = sede_element.find_next('a')['title']
            speaker = speaker_elements.find_next('a')['title']
            winner = winner_elements.find_next('a')['title']
            if singers_elements is not None:
                #singers = singers_elements.get_text(strip=True)
                singers = singers_elements.find_next('td').get_text(strip=True)
                # Esegui altre operazioni con il testo estratto
            else:
                print(year," - ","Elemento non trovato.")
            # Aggiungi il titolo della sede e l'anno alla lista dei dati
            data_list.append({'Anno': year,'Periodo': period, 'Sede': sede,'Presentatore': speaker,'Partecipanti': singers,'Vincitore': winner})

# Ora hai una lista di dizionari, ognuno contenente l'anno e la sede estratta da una pagina
# Puoi convertire questa lista in un DataFrame pandas se lo desideri
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Funzione per estrarre la tabella dalla pagina HTML
def scrape_table(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Trova il tag <span> con l'id "Classifica_finale" o "Classifica,partecipanti"
        span_tag = soup.find('span', id=['Classifica_finale', 'Classifica,partecipanti','Classifica, canzoni e cantanti','Classifica,_canzoni_e_cantanti','Classifica,_canzoni,_cantanti'])
        if span_tag:
            # Trova la tabella successiva al tag <span>
            table = span_tag.find_next('table')
            if table:
                # Leggi la tabella e crea il DataFrame
                df = pd.read_html(str(table))[-1]  # La funzione restituisce una lista di DataFrame, prendiamo il primo
                return df
            else:
                print(year,"Nessuna tabella trovata dopo il tag 'Classifica finale' o 'Classifica,partecipanti'.")
        else:
            print(year,"Nessun tag 'Classifica finale' o 'Classifica,partecipanti' trovato sulla pagina.")
    else:
        print("Errore durante il recupero della pagina:", response.status_code)

# Lista per memorizzare i DataFrame di ogni anno
dataframes_by_year = {}

# Ciclo attraverso gli anni da 2019 a 2024
for year in range(1951, 2024):
    # Costruisci l'URL per l'anno corrente
    url = f'https://it.wikipedia.org/wiki/Festival_di_Sanremo_{year}'

    # Esegui la funzione per estrarre la tabella e creare il DataFrame per l'anno corrente
    df = scrape_table(url)
    if df is not None:
        # Aggiungi il DataFrame all'elenco dei DataFrame per l'anno corrente
        dataframes_by_year[year] = df

# Ora hai un dizionario di DataFrame, dove ogni DataFrame rappresenta i dati di un anno specifico.
# Puoi accedere ai DataFrame per ciascun anno utilizzando l'anno come chiave nel dizionario.
```

```python
def get_song_info(song_name):
    # Esegui la ricerca per il nome della canzone corrente
    results = sp.search(q=song_name, limit=1, type='track')
    # Inizializza le variabili per salvare le informazioni
    artist_name = None
    year = None
    duration_minutes = None
    popularity = None
    album_image = None
    # Analizza i risultati della ricerca
    for track in results['tracks']['items']:
        # Ottieni l'artista
        for artist in track['artists']:
            artist_name = artist['name']
            break
        # Ottieni l'anno di rilascio dell'album
        release_date = sp.album(track['album']['id'])['release_date']
        year = release_date.split('-')[0]  # Ottieni solo l'anno dalla data
        # Ottieni il minutaggio della traccia in millisecondi e convertilo in minuti
        duration_ms = track['duration_ms']
        duration_minutes = duration_ms / 1000 / 60
        # Ottieni la popolarità della traccia
        popularity = track['popularity']
        # Ottieni l'URL dell'immagine dell'album
        album_image = track['album']['images'][0]['url']
        break  # Esci dal ciclo delle tracce dopo aver trovato l'artista per la traccia corrente
    # Restituisci le informazioni della canzone
    return artist_name, year, duration_minutes, popularity, album_image
```
